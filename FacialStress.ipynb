```python
import os

base_path = "/content/archive (6) (1)/Chips_Thermal_Face_Dataset/images"
image_files = os.listdir(base_path)

print(image_files[:5])
```

```python
import cv2
import numpy as np
import pandas as pd

data = []

for img_name in image_files:
    img_path = os.path.join(base_path, img_name)

    img = cv2.imread(img_path)
    if img is None:
        continue

    img = cv2.resize(img, (128, 128))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 🔹 Feature 1: Face Temperature (mean intensity)
    face_temp = np.mean(gray)

    # 🔹 Feature 2: Heatmap Intensity (max intensity)
    heatmap_intensity = np.max(gray)

    # 🔹 Feature 3: Facial Fatigue (variance - lower variance = fatigue)
    fatigue = np.var(gray)

    # 🔹 Output (Facial Stress - YOU DEFINE)
    # Simple logic (you can improve later)
    stress = (face_temp * 0.4) + (heatmap_intensity * 0.4) - (fatigue * 0.2)

    data.append([face_temp, fatigue, heatmap_intensity, stress])

# Convert to DataFrame
df = pd.DataFrame(data, columns=[
    "face_temperature",
    "facial_fatigue",
    "heatmap_intensity",
    "facial_stress"
])

print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import joblib

# Inputs & Output
X = df[["face_temperature", "facial_fatigue", "heatmap_intensity"]]
y = df["facial_stress"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestRegressor()
model.fit(X_train, y_train)

# Save model
joblib.dump(model, "facial_stress_model.joblib")

print("✅ Model trained & saved!")
```

```python
sample = [[120, 500, 200]]  # example input
prediction = model.predict(sample)

print("Predicted Facial Stress:", prediction)
```
