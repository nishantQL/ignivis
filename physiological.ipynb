```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/physiological_1.csv")

# Check data
print(df.head())
```

```python
df = df.dropna()
features = ['Body Temperature', 'Heart Rate', 'Respiratory Rate', 'Oxygen Saturation']
X = df[features]
```

```python
# Create score manually (your innovation)
df['physiological_score'] = (
    0.3 * ((df['Body Temperature'] - 36) / (40 - 36) * 100) +
    0.2 * ((df['Heart Rate'] - 60) / (140 - 60) * 100) +
    0.2 * ((df['Respiratory Rate'] - 12) / (30 - 12) * 100) +
    0.1 * ((100 - df['Oxygen Saturation']) * 2)
)

# Limit score between 0–100
df['physiological_score'] = df['physiological_score'].clip(0, 100)

y = df['physiological_score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestRegressor()
model.fit(X_train, y_train)

# Check accuracy
score = model.score(X_test, y_test)
print("Model Accuracy:", score)
```

```python
import joblib

# Save model
joblib.dump(model, "heat_stress_model.joblib")

print("Model saved successfully!")
```

```python
import joblib

# Load model
model = joblib.load("heat_stress_model.joblib")

# Test prediction
sample = [[38.5, 110, 22, 95]]
prediction = model.predict(sample)

print("Predicted Score:", prediction[0])
```
